# 3.Loading the Data

```python
# Lets define a function that reads data

def read(url,csv):
    response=requests.get(url)
    with open(csv, mode ='wb') as file:
        file.write(response.content)
    
    
```

```python
# Fetching the data programatically.

url="https://video.udacity-data.com/topher/2020/October/5f91cf38_201902-fordgobike-tripdata/201902-fordgobike-tripdata.csv"
csv = '201902-fordgobike-tripdata.csv'

# Calling the function we had created
read(url,csv)
```

## Defining Issue Two :
Correct columns with wrong datatypes.

## Code :

```python
# Lets convert `bike_id`, `start_station_id`,  `end_station_id` into string Datatypes

clean_df['start_station_id'] =clean_df['start_station_id'].astype(str)
clean_df['end_station_id'] = clean_df['end_station_id'].astype(str)
clean_df['bike_id'] = clean_df['bike_id'].astype(str)

# Lets convert member birth year from float to int
clean_df['member_birth_year']= clean_df['member_birth_year'].astype(int)

# Lets convert duration_sec to float
clean_df['duration_sec'] = clean_df['duration_sec'].astype(float)


# Lets convert start_time and end_time to data time format
clean_df['start_time'] = pd.to_datetime(clean_df['start_time'])
clean_df['end_time'] = pd.to_datetime(clean_df['end_time'])

##convert start_time to hours, days, month and year
clean_df['start_day'] = clean_df['start_time'].dt.day_name()
clean_df['start_month'] = clean_df['start_time'].dt.month_name()

# Lets convert end_time to hours, days, month and year

clean_df['end_day'] = clean_df['end_time'].dt.day_name()
clean_df['end_month'] = clean_df['end_time'].dt.month_name()

#bike['end_year'] = bike['end_time'].dt.year

# Lets calculate the age of the participants when this trip started. 
clean_df['age']=(2019 - clean_df.member_birth_year)
```

```python
# Lets create a function that gives the xlabel, ylabel and title to any graph
def plot(Xlabel, ylabel, title):
    plt.title(title)
    plt.xlabel(Xlabel)
    plt.ylabel(ylabel)
    
```

```python
bins= np.arange(61, clean_df.duration_sec.max()+200, 200)
plt.figure(figsize=(10, 8))
plt.hist(data = clean_df, x = 'duration_sec', bins=bins)
plt.xlim(0, 6000)
plot('duration (sec)',' Count', 'Distribution of trip duration')
plt.show()

```

> Duration in seconds has a long-tailed distribution and was skewed to right. This implies that the highest number of participants took lesser time (seconds) to complete the trip, while few participants took longer time to complete the trip.

```python
# Let's put it on a log scale instead
bins= 10**np.arange(1, 5+0.1, 0.1)
#set figure size
plt.figure(figsize=(15, 5))
plt.hist(data = clean_df, x = 'duration_sec', bins=bins)
plt.xscale('log')
plot('duration (sec)',' Count', 'Distribution of Log-transformed trip duration')
plt.xticks([50, 100, 250, 500, 1000, 3000, 6000],[50, 100,250, 500, 1500, 3000, 6000])
plt.show()
```

 When we plotted on a log-scale, the distribution of the duration of time looks unimodal with peak between 400 and 1000.

## Bivariate Exploration

> In this section, I investigated the relationships between pairs of variables the clean data. These variables were introduced in the univariate analysis and included trip duration, age, gender, user type and trip start day. 

I'll look at the relationship between trip duration and age

```python
#A scatter plot of age and duration_sec with log transform on both age and duration
plt.figure(figsize=(10,5))
bike_sub = clean_df.sample(1000, replace=False)
def log_trans(x, inverse=False):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)
sns.regplot(bike_sub['age'].apply(log_trans), bike_sub['duration_sec'].apply(log_trans),  scatter_kws={'alpha':1/5}, x_jitter=0.2)
# Lets add xlabel, ylabel and title
plot('Age', 'Trip duration (sec)', 'Scatter plot of log-age vs. log-duration')
plt.show()
```

> The scatter plot shows that there is a moderate linear relationship between age and trip duration after applying log transformation on the two variables.

- Most of the bike riders, who participated in this trip were aged between 18 and 50 year
- The number of males who participated in the Ford GoBike trips were more than females.
- Most of the participants in the Ford GoBike trips were subscribers and not customers.
- There is a moderate linear relationship between age and trip duration after log transformation of the two variables.
- Subscribers took lesser time to complete the trip than customers.
- Male participants appear to complete the trip faster than females.
- Participants took more time to complete the trip on saturdays than other days of the week.
- There were more males than females both in the customer and subscriber category. 
- Those who shared bike took slightly lesser time than those who did not share the bikes.
- Regardless of the user type, male participants took lesser time to complete the trip than females.
